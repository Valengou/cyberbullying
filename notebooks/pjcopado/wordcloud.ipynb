```python
text_0 = ''.join(df[df['target']==0]['text'])
```

```python
text_1 = ''.join(df[df['target']==1]['text'])
```

```python
def remove_short_words(text, n=2):
    return ' '.join([word for word in text.split() if len(word)>n])
```

```python
text_0 = remove_short_words(text_0)
text_1 = remove_short_words(text_1)
```

```python
text_1[:1000]
```

```python
wordcloud_1 = WordCloud(max_words=100).generate(text_1)
plt.figure(figsize=(12,6))
plt.imshow(wordcloud_1)
plt.axis("off")
plt.show()
```

```python
wordcloud_1.words_
```
